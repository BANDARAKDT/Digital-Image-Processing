5.	Rotate the given image (image_1.jpg)  by 90 degrees and display.

```python
img1=cv2.imread('image_01.jpg')
(h,w)=img1.shape[:2]
center = (w//2,h//2)
matrix= cv2.getRotationMatrix2D(center,90,1.0)
rotated_image= cv2.warpAffine(img1,matrix,(w,h))
cv2.imwrite('Rotated_image.jpg',rotated_image)
cv2_imshow(rotated_image)
```

10.

a)	Draw a rectangle image as a 200 x 200 NumPy array.
b)	Draw a white filled rectangle on the black image using CV2. (Use the (25, 25) as the Top-left corner and the (275, 275) as the Bottom-right corner of the rectangle.)
c)	Draw a rectangle image as a 200 x 200 NumPy array.
d)	Draw a filled white circle on the black image according to the following properties.
•	The center of the circle : x=100, y=100
•	The radius of the circle: 100 pixels).
•	Color:  White
Perform the Bitwise Operations (AND, OR, XOR, NOT) using the drawn images above ‘b’ and ‘d’


```python
import numpy as np

import matplotlib.pyplot as plt
# 10a. Create a Black 200x200 Image
rect_image = np.zeros((200, 200), dtype=np.uint8)

# 10b. Draw a White Filled Rectangle
cv2.rectangle(rect_image, (25, 25), (175, 175), 255, -1)
plt.imshow(rect_image, cmap="gray")
plt.title("White Rectangle on Black")
plt.axis("off")
plt.show()

# 10c. Create Another 200x200 Black Image
circle_image = np.zeros((200, 200), dtype=np.uint8)

# 10d. Draw a White Filled Circle
cv2.circle(circle_image, (100, 100), 100, 255, -1)
plt.imshow(circle_image, cmap="gray")
plt.title("White Circle on Black")
plt.axis("off")
plt.show()

# Perform Bitwise Operations
bitwise_and = cv2.bitwise_and(rect_image, circle_image)
bitwise_or = cv2.bitwise_or(rect_image, circle_image)
bitwise_xor = cv2.bitwise_xor(rect_image, circle_image)
bitwise_not_rect = cv2.bitwise_not(rect_image)

# Display Bitwise Operations
plt.figure(figsize=(10, 10))

plt.subplot(221)
plt.imshow(bitwise_and, cmap="gray")
plt.title("Bitwise AND")

plt.subplot(222)
plt.imshow(bitwise_or, cmap="gray")
plt.title("Bitwise OR")

plt.subplot(223)
plt.imshow(bitwise_xor, cmap="gray")
plt.title("Bitwise XOR")

plt.subplot(224)
plt.imshow(bitwise_not_rect, cmap="gray")
plt.title("Bitwise NOT (Rectangle)")

plt.show()
```
