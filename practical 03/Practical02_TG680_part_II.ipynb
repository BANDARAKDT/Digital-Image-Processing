**9.merge the separate color channels to the single image**

```python
import cv2
from google.colab.patches import cv2_imshow  # Import cv2_imshow for Google Colab

image = cv2.imread("opencv_logo.png")
B,G,R = cv2.split(image)
merged_opencv_logo = cv2.merge([B, G, R])
cv2.imwrite("Merged_opencv_logo.png", merged_opencv_logo)
merged_img = cv2.imread("Merged_opencv_logo.png")
cv2_imshow(merged_img)
```

**12.Addition and subscription**

we need similar size of images to add images. find thesuitable images on the internet and performthe following arithmetics opertations and write the final image to yhe googlecolabes .1.Addition 2. substraction

```python
import cv2
from google.colab.patches import cv2_imshow  # Import cv2_imshow for Google Colab


Image1 = cv2.imread('image_01.jpg')
Image2 = cv2.imread('image_02.jpg')
result = cv2.add(Image2,Image1)
cv2_imshow(result)
```

```python
Image1 = cv2.imread('image_01.jpg')
Image2 = cv2.imread('image_02.jpg')
result = cv2.subtract(Image1, Image2)
cv2_imshow(result)
```

**14. draw shapes**

```python
import numpy as np
# Draw a rectangle
rectangle = np.zeros((300,300), dtype="uint8")
cv2.rectangle(rectangle,(25,25),(275,275),255,-1)   # -1 is use to completely fill the squares
cv2_imshow(rectangle)

```

```python
#draw a circle
circle= np.zeros((300,300), dtype="uint8")
cv2.circle(circle,(150,150),150,255,-1)
cv2_imshow(circle)
```

```python
# betwise AND
import numpy as np
# Draw a rectangle
rectangle = np.zeros((300,300), dtype="uint8")
cv2.rectangle(rectangle,(25,25),(275,275),255,-1)   # -1 is use to completely fill the squares
cv2_imshow(rectangle)
#draw a circle
circle= np.zeros((300,300), dtype="uint8")
cv2.circle(circle,(150,150),150,255,-1)
cv2_imshow(circle)

bitwiseAND =cv2.bitwise_and(rectangle,circle)
cv2_imshow(bitwiseAND)
```
